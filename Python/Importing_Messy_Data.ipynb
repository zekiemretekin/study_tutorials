## Importing Messy Data

__Import__ the cars Dataset from the messy csv-file __cars_raw.csv__ into a Pandas DataFrame. Use appropriate __parameters__ in the __pd.read_csv()__ method to bring the DataFrame into a clean format. __Columns__ should have the following __labels__: Finally, __save__ and __export__ the dataset as new csv-file (__cars_imp.csv__).

```python
labels = ['mpg',
 'cylinders',
 'displacement',
 'horsepower',
 'weight',
 'acceleration',
 'model year',
 'origin',
 'name']
```

```python
import pandas as pd
```

```python
cars=pd.read_csv("cars_raw.csv", skiprows=2, skipfooter=1, header=None, names=labels)
```

```python
cars.to_csv("cars_imp.csv", index=False)
```
