### Plotting Data with Matplotlib

```python
import pandas as pd
```

```python
cars = pd.read_csv("cars.csv")
```

73. __Import__ matplotlib.pyplot (as plt) and set the __style__ to __"seaborn"__! __Fill in the gaps!__

```python
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

74. __Create__ the graph below (lineplot for all numerical columns)!

```python
cars.plot(kind="line",subplots=True, figsize=(15,15))
plt.show()
```

75. __Create__ the graph below (Histogramm for the __"mpg"__ column with 40 bins)!

```python
cars.mpg.hist(figsize = (12,8), bins = 40)
plt.title="MPG"
plt.xlabel="mpg"
plt.ylabel="frequency"
plt.show()
```

76. __Create__ the graph below (Scatterplot with __horsepower__ and __mpg__)! __Fill in the gaps!__

```python
cars.plot(kind = "scatter", x = "horsepower", y = "mpg", figsize = (12,8), c = "cylinders", marker = "x", colormap = "viridis")
plt.title("MPG vs. Horsepower", fontsize = 18)
plt.xlabel("horsepower", fontsize = 15)
plt.ylabel("mpg", fontsize = 15)
plt.show()
```
